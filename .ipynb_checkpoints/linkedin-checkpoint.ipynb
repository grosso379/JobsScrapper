```python
import csv
from datetime import datetime
import requests
from bs4 import BeautifulSoup
```

```python
def getLinkedinUrl(position, location):
    position = position.replace(" ", "%20")
    location = location.replace(" ", '%20')
    return f'https://www.linkedin.com/jobs/search/?keywords={position}&location={location}'
```

```python
linkedinUrl = getLinkedinUrl('Software Engineer', 'Argentina')
linkedinUrl
```

## Extract raw url

```python
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
response = requests.get(linkedinUrl, headers=headers, timeout=5)
```

```python
linkedinSoup = BeautifulSoup(response.text, 'html')
linkedinSoup
```

```python
linkedinJobCards = linkedinSoup.find_all('div', {"class":"hide-on-mobile"})
linkedinJobCards
```
