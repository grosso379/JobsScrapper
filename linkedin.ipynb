```python
import csv
from datetime import datetime
import requests
from bs4 import BeautifulSoup
```

```python
def getLinkedinUrl(position, location):
    position = position.replace(" ", "%20")
    location = location.replace(" ", '%20')
    return f'https://www.linkedin.com/jobs/search/?keywords={position}&location={location}'
```

```python
linkedinUrl = getLinkedinUrl('Software Engineer', 'Argentina')
linkedinUrl
```

## Extract raw url

```python
headers = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
response = requests.get(linkedinUrl, headers=headers)
```

```python
linkedinSoup = BeautifulSoup(response.text, 'html')
```

```python
linkedinJobCards = linkedinSoup.find_all('div', {"class":"job-search-card"})
```

```python
prototype = linkedinJobCards[0]
link = prototype.a['href']
```

```python
jobResponse = requests.get(link, headers=headers)
jobSoup = BeautifulSoup(jobResponse.text, 'html')
```

```python
jobTitle = jobSoup.h1.get_text(strip=True)
jobCompany = jobSoup.find('a', 'topcard__org-name-link').get_text(strip=True)
```
